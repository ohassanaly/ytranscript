```python
from utils import get_youtube_video_id, retrieve_video_transcript
import json
from tqdm import tqdm
import time
import random
import logging
```

```python
with open("c3ai_videos.json", "r", encoding="utf-8") as f:
    data = json.load(f)

out_path = "c3ai_transcripts.jsonl"

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s %(levelname)s [%(name)s] %(message)s",
    filename="full_extract_c3ai.log",
    filemode="a",  # append
)

logger = logging.getLogger(__name__)
```

```python
result = {}
for item in tqdm(data):
    title=item["title"]
    url=item["url"]
    try:
        record = {
            "url": url,
            "title": title,
            "transcript": retrieve_video_transcript(url),
        }
    except Exception:
        logger.exception("Failed processing url=%s", url)
        record = {"url": url, "title": title, "transcript": ""}
    #saving transcripts one by one
    with open(out_path, "a", encoding="utf-8") as f:
        f.write(json.dumps(record, ensure_ascii=False) + "\n")
        f.flush()

    time.sleep(random.uniform(2, 5)) #presumably, the API is rate-limited to 5 requests per 10 seconds
```
