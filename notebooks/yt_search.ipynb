In this notebook, we retrieve all the video url of a given youtube channel 

```python
from googleapiclient.discovery import build
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
youtube = build("youtube", "v3", developerKey=os.getenv("YOUTUBE_API_KEY"))
```

```python
#https://developers.google.com/youtube/v3/docs/channels/list

#Step 1: Get the channel’s Uploads playlist ID
request = youtube.channels().list(
    part="contentDetails",
    forHandle="@snowflakedevelopers",
)
response = request.execute()
uploads_playlist_id = response["items"][0]["contentDetails"]["relatedPlaylists"]["uploads"]

#Step 2: List videos from the uploads playlist

videos = {}

request = youtube.playlistItems().list(
    part="snippet",
    playlistId=uploads_playlist_id,
    maxResults=50
)

while request:
    response = request.execute()

    for item in response["items"]:
        title = item["snippet"]["title"]
        video_id = item["snippet"]["resourceId"]["videoId"]
        url = f"https://www.youtube.com/watch?v={video_id}"

        videos[title] = url

    request = youtube.playlistItems().list_next(request, response)
```

```python
#save the result (list of videos) 
import json

with open("snowflakedevelopers_videos.json", "w", encoding="utf-8") as f:
    json.dump(videos, f, indent=2, ensure_ascii=False)
```
