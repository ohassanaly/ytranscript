```python
import json

with open("snowflakedevelopers_videos.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
# Flattening logic
flattened_data = [{"title": title, "url": url} for title, url in data.items()]
```

```python
with open("snowflake_flat.json", "w", encoding="utf-8") as f:
    json.dump(flattened_data, f, indent=2, ensure_ascii=False)
```
